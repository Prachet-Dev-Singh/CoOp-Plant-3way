```python
import os

root_dir = 'CoOp/data'
plant_disease_dir = os.path.join(root_dir, 'plant_disease_3way')

print(f"Root directory exists: {os.path.exists(root_dir)}")
print(f"Plant disease directory exists: {os.path.exists(plant_disease_dir)}")

if os.path.exists(plant_disease_dir):
    for split in ['Train', 'Validation', 'Test']:
        split_dir = os.path.join(plant_disease_dir, split)
        print(f"\n{split} directory exists: {os.path.exists(split_dir)}")
        if os.path.exists(split_dir):
            print(f"Contents of {split} directory:")
            for item in os.listdir(split_dir):
                item_path = os.path.join(split_dir, item)
                print(f"- {item} (directory: {os.path.isdir(item_path)}, file: {os.path.isfile(item_path)})")
else:
    print("Plant disease directory does not exist.")
```
